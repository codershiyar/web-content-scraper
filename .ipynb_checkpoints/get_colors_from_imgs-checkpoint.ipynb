```python
import extcolors
from collections import Counter
import os
import concurrent.futures
import matplotlib.pyplot as plt

list_colors = []
img_dir = "imgs/"
num_threads = 4

def most_common_colors(img_file):
    img_file = img_dir + img_file
    colors = extcolors.extract_from_path(img_file, tolerance=12, limit=12)[0]
    color_counts = Counter(dict(colors)).most_common(2)
    print(f" foto {img_file} is done")
    return color_counts

def main():
    img_files = os.listdir(img_dir)
    with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
        futures = [executor.submit(most_common_colors, img_file) for img_file in img_files]
        for future in concurrent.futures.as_completed(futures):
            result = future.result()
            list_colors.append(result)
def draw_dashboard(list_colors):
    most_common_colors = sorted(list_colors, key=lambda x: x[0][1], reverse=True)[:3]
    most_3common_colors = []
    height = []
    for i, color in enumerate(most_common_colors):
        print(f"De #{i+1} meest voorkomende kleur is: {color[0][0]}")
        height.append(color[0][1])
        color = color[0][0]
        most_3common_colors.append((color[0]/255,color[1]/255, color[2]/255 ))
    
    print(most_3common_colors)   

    left = [1, 2, 3]
# labels for bars
    tick_label = ['Kleur1', 'Kleur2', 'Kleur3']
# plotting a bar chart
    plt.bar(left, height, tick_label=tick_label, width=0.8, color=most_3common_colors)
# plot title
    plt.title('De meeste 3 populaire productkleuren.')
# function to show the plot
    plt.show()    

if __name__ == '__main__':
    main()
    draw_dashboard(list_colors)
    


```
